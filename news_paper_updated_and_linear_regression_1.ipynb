```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import statsmodels.formula.api as smf
```

```python
df=pd.read_csv("/content/NewspaperData.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.displot(df['daily'])
```

```python
sns.regplot(x="daily",y="sunday",data=df)
```

```python
plt.plot(df.daily,df.sunday,'bo')
plt.xlabel("daily")
plt.ylabel("sunday")
plt.title("daily vs sunday")
plt.show()
```

```python
df.corr(numeric_only=True)
```

```python
numerical_data=df.select_dtypes(include=np.number)
numerical_data.corr()
```

```python
model=smf.ols("sunday~daily",data=df).fit()
model.summary()
```

```python
model.params
```

```python
(model.rsquared,model.rsquared_adj)
```

```python
y_pred=model.predict(df['daily'])
y_pred
```

```python
newData=pd.Series([300,400,900])
data_pred=pd.DataFrame(newData,columns=['daily'])
data_pred
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import statsmodels.formula.api as smf
```

```python
data=pd.read_csv("wc-at.csv")
data
```

```python
sns.displot(data['Waist'])
plt.show()
```

```python
plt.boxplot(data.AT)
```

```python
sns.regplot(x='Waist',y='AT',data=data)
```

```python
data.corr()
```

```python
model=smf.ols("AT~Waist",data=data).fit()
model.summary()
```

```python
model.rsquared
```

```python
model.rsquared_adj
```

```python
model.resid
```

```python
data.shape
```

```python
plt.hist(data.Waist)
plt.show()
```

```python
y_pred=model.predict(data['Waist'])
y_pred
```
